## Find image features using FAST

Grayscale image

```python
img_grayscale = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
```

```python
detector = cv2.FastFeatureDetector_create()
detector.setThreshold(70)

# find and draw the keypoints
kp = detector.detect(img_grayscale,None)
img_detect=cv2.drawKeypoints(img_grayscale, kp,outImage=None,flags=4)
print "Number of corner features: " + str(len(kp))
```

```python
img_detect=img_rgb.copy()
```

```python
for circle in kp:
    centroid=tuple([int(x) for x in circle.pt])
    cv2.circle(img_detect,centroid,20,(255,0,0),3)
```

Color features so we can see them

Show output feature radius circles

```python
plt.figure(figsize=(20,10))
plt.imshow(img_detect),plt.show()
```

```python
cv2.imwrite("Figures/SurfLizard.jpg",img_detect)
```
